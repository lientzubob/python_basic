```python
data = pd.DataFrame({'name': ['aa', 'AA', 'bb', 'BB', 'cc', 'CC'],
                     'number' : [1, 2, 3, 4, 5, 6]})
data
```

```python
lower_name = data['name'].str.lower()
lower_name
```

```python
data['name'].str.upper()
```

```python
age = {
    'aa': 10,
    'bb': 20,
    'cc': 30
}
data['age'] = lower_name.map(age)
data
```

```python
data['name'].map(lambda x: age[x.lower()])
```
